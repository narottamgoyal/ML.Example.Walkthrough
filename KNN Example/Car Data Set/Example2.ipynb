# Column Label encoding

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
dataset['buying'].value_counts()
```

```python
label_encoder = preprocessing.LabelEncoder()
dataset['buying_original_df'] = dataset[['buying']].copy()
dataset['buying']= label_encoder.fit_transform(dataset['buying']) 
# dataset = pd.concat([dataset, buying_original_df])
dataset['buying'].value_counts()
```

```python
dataset[['buying_original_df' , 'buying']]
```

```python
dataset['maint']= label_encoder.fit_transform(dataset['maint']) 
dataset['door']= label_encoder.fit_transform(dataset['door']) 
dataset['persons']= label_encoder.fit_transform(dataset['persons']) 
dataset['lug_boot']= label_encoder.fit_transform(dataset['lug_boot']) 
dataset['safety']= label_encoder.fit_transform(dataset['safety'])
dataset['class']= label_encoder.fit_transform(dataset['class']) 
dataset
```

```python
dataset.info()
```

# Split

```python
x = dataset.iloc[:,: len(dataset.columns)-1]
y = dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.3)
```

```python
x_train, y_train
```

# Training

```python
from sklearn.neighbors import KNeighborsClassifier
knnclassifier = KNeighborsClassifier(n_neighbors=5)
```

```python
knnclassifier.fit(x_train, y_train)
```

```python
acc = knnclassifier.score(x_test, y_test)
print(acc)
```

```python
import numpy as np
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
prediction = knnclassifier.predict(x_test)
```
