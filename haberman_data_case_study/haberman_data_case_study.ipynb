```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

file='haberman.csv'
haberman = pd.read_csv(file)
```

```python
# status (class attribute): 
# 1 = the patient survived 5 years or longer
# 2 = the patient died within 5 year
```

```python
haberman.plot(kind='scatter', x = "age", y='nodes')
```

```python
sns.set_style()#'whitegrid')#whitegrid
sns.FacetGrid(haberman, hue='status', size=4) \
                .map(plt.scatter, "age", 'nodes') \
                .add_legend()
plt.show()
```

```python
sns.set_style()#'whitegrid')#whitegrid
sns.FacetGrid(haberman, hue='status', size=8) \
                .map(plt.scatter, "age", 'nodes') \
                .add_legend()
plt.show()
```

```python
plt.close()
sns.set_style('whitegrid')#whitegrid
sns.pairplot(haberman, hue='status', size=3)
plt.show()
```

```python
import numpy as np
one = haberman.loc[haberman["status"] == "1"];
two = haberman.loc[haberman["status"] == "2"];
#print(iris_setosa["petal_length"])
plt.plot(one["age"], np.zeros_like(one['nodes']), 'o')
plt.plot(two["age"], np.zeros_like(two['nodes']), 'o')

plt.show()
```

```python
sns.FacetGrid(haberman, hue="status", size=6) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

```python
plt.figure(1,figsize=(14,4))
plt.subplot(1,3,1)
sns.boxplot(data=haberman,x='status',y='age')
plt.subplot(1,3,2)
sns.boxplot(data=haberman,x='status',y='year')
plt.subplot(1,3,3)
sns.boxplot(data=haberman,x='status',y='nodes')
```
