```python
pd.concat([df_original['MSZoning'].value_counts()/df_original.shape[0] * 100, df['MSZoning'].value_counts()/df.shape[0] * 100],
             axis=1,
         keys=['MSZoning_original', 'MSZoning_clean'])
```

```python
def cat_var_dis(var):
    return pd.concat([df_original[var].value_counts()/df_original.shape[0] * 100, df[var].value_counts()/df.shape[0] * 100],
             axis=1,
         keys=[var+'_original', var+'_clean'])
```

```python
cat_var_dis('MSZoning')
```
