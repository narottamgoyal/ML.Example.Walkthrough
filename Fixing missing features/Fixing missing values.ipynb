## Mean, Median, mode 

```python
df = df_original.copy()
```

```python
df[df.isnull().any(axis=1)].head(2)
```

```python
missing_num_var = [var for var in df.select_dtypes(include=['int64', 'float64']).columns if df[var].isnull().sum() > 0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i +1)
    sns.distplot(df[var], bins=20, kde_kws={'linewidth': 5, 'color': '#DC143C'})
```

```python
df[missing_num_var].isnull().sum().sum()
```

```python
df_num_mean = df[missing_num_var].fillna(df[missing_num_var].mean())
df_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i +1)
    sns.distplot(df_num_mean[var], bins=20, kde_kws={'linewidth': 5, 'color': 'green'})
    sns.distplot(df_original[var], bins=20, kde_kws={'linewidth': 5, 'color': 'red'})    
```

```python
df_num_median = df[missing_num_var].fillna(df[missing_num_var].median())
df_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i +1)
    sns.distplot(df_num_mean[var], bins=20, kde_kws={'linewidth': 5, 'color': 'green'})
    sns.distplot(df_original[var], bins=20, kde_kws={'linewidth': 5, 'color': 'red'})    
    sns.distplot(df_num_median[var], bins=20, kde_kws={'linewidth': 5, 'color': 'black'})
```

# outlier

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3, 1, 1)
    sns.boxplot(df[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(df_num_mean[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(df_num_median[var])
```

# using mean

```python
df_copy_mean = df.copy()
num_var_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig' , 'Exterior2nd', 'KitchenQual']
for cat_var, num_var in zip(cat_vars, num_var_miss):
    for var in df[cat_var].unique():
        df_copy_mean.update(df[df.loc[:, cat_var] == var][num_var].replace(np.nan, df[df.loc[:, cat_var] == var][num_var].mean()))
```

```python
df_copy_mean[missing_num_var].isnull().sum()
```

# using median

```python
df_copy_median = df.copy()
num_var_miss = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig' , 'Exterior2nd', 'KitchenQual']
for cat_var, num_var in zip(cat_vars, num_var_miss):
    for var in df[cat_var].unique():
        df_copy_median.update(df[df.loc[:, cat_var] == var][num_var].replace(np.nan, df[df.loc[:, cat_var] == var][num_var].median()))
```

```python
df_copy_median[missing_num_var].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2, i +1)
    sns.distplot(df_copy_mean[var], bins=20, kde_kws={'linewidth': 5, 'color': 'green'}, label='mean')
    sns.distplot(df_copy_median[var], bins=20, kde_kws={'linewidth': 5, 'color': 'red'}, label='median')
    plt.legend()
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3, 1, 1)
    sns.boxplot(df[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(df_copy_mean[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(df_copy_median[var])
```
